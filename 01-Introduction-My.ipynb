```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train, y_train)
```

```python
import numpy as np
```

```python
X_new = np.array([[5, 2.9, 1, 0.2]])
```

```python
print("Форма массива X_new: {}".format(X_new.shape))
```

```python
prediction = knn.predict(X_new)
print("Прогноз: {}".format(prediction))
print("Спрогнозированная метка: {}".format(
                                iris_dataset['target_names'][prediction]))
```

```python
y_pred = knn.predict(X_test)
print("Прогнозы для тестового набра:\n {}".format(y_pred))
```

```python
print("Правильность на тестовом наборе: {:.3}".format(np.mean(y_pred == y_test)))
```

```python
print("Правильность на тестовом наборе: {:.3}".format(knn.score(X_test, y_test)))
```
